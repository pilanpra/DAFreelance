```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def extract_text_from_html(html, class_name):
    soup = BeautifulSoup(html, 'html.parser')
    # Find elements by class name
    elements = soup.find_all(class_=class_name)
    if elements:
        return ' '.join(element.get_text(strip=True) for element in elements)
    else:
        # If class name not found, try finding by page header value
        element = soup.find(string=lambda text: text.strip().lower() == class_name.lower())
        if element:
            return element.find_next().get_text(strip=True)
    return "Abstract not found on the page."

def get_abstract_or_pdf_link(row):
    url = row['Link']
    page_header = row['Page_Header']
    try:
        response = requests.get(url)
        response.raise_for_status()  

        if isinstance(page_header, str) and page_header.lower() == 'pdf':
            # Return PDF download link
            return url
        else:
            # Extract abstract text based on class name or page header value
            abstract_text = extract_text_from_html(response.content, str(page_header))
            return abstract_text

    except requests.exceptions.RequestException as e:
        return f"Error: {e}"
```

```python
import pandas as pd
from urllib.parse import urlparse

def get_website_domain(link):
    try:
        if isinstance(link, str):  # Check if link is a string
            parsed_url = urlparse(link)
            return parsed_url.netloc
        else:
            return "Invalid URL"
    except Exception as e:
        return f"Error: {e}"

master_df['Website'] = master_df['Link'].apply(get_website_domain)

grouped_df = master_df.groupby('Website').size().reset_index(name='Number of Articles')
grouped_df = grouped_df.sort_values(by='Number of Articles', ascending=False)

# for website, group in grouped_df:
#     print(f"Website: {website}")
#     print(group)
grouped_df.head(50)
```
